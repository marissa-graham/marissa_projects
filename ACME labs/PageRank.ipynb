```python
from scipy.sparse import dok_matrix
import numpy as np
from scipy import linalg as la


def to_matrix(datafile, n ):
    
    ''' Return the nxn adjacency matrix described by datafile.
    INPUTS:
    datafile - A .txt file describing a directed graph. Lines
    describing edges should have the form '<from node>\t<to node>\n'. The file may also
    include comments.
    n - The number of nodes in the graph described by datafile

    RETURN:
    Return a SciPy sparse `dok_matrix'.
    '''
    A = dok_matrix((n,n))
    with open(datafile,'r') as myfile:
        for line in myfile:
            a = line.strip().split()
            if a[0] != '#':
                try:
                    A[int(a[0]),int(a[1])]=1
                except:
                    raise ValueError
    return A

def iter_solve(data, datasize, N=None, d=.85, tol=1E-5,maxiters=500):

    '''
    Return the page ranks of the network described by `data`.
    
    INPUTS:
    data - A NumPy array representing the adjacency matrix of a directed graph
    datasize - size of original matrix
    N - Restrict the computation to the first `N` nodes of the graph. Defaults to N=None; in this case, the entire matrix is used.
    d - The damping factor, a float between 0 and 1. Defaults to .85.
    tol - Stop iterating when the change in approximations to the solution is less than `tol'. Defaults to 1E-5.

    OUTPUTS:
    Iterate through the PageRank algorithm until the error is less than
    `tol'. Return the approximation to the steady state.
    '''
    if N:
        A = data[:N,:N]
        n = N
    else:
        A = data
        n = datasize
        
    Am = A*np.eye(n)
    for i in xrange(n):
        if np.sum(Am,axis=1)[i] == 0:
            Am[i,:] += 1
            
    D = np.sum(Am,axis=1)
    K = Am.T/D
    
    
    
    iters = 0
    error = 1
    p0 = np.ones(n)/n
    to_add = (1-d)*np.ones(n)/n
    pk = d*np.dot(K,p0) + to_add
    while error >= tol and iters < 500:
        iters += 1
        pnext = d*np.dot(K,pk) + to_add
        error = la.norm(pnext-pk)
        pk = pnext

    return pk
        
def eig_solve(data, datasize, N=None, d=.85):
    '''
    Return the page ranks of the network described by `data`.

    INPUTS:
    data - A NumPy array representing the adjacency matrix of a directed graph
    datasize - size of original matrix
    N - Restrict the computation to the first `N` nodes of the graph. Defaults to N=None; in this case, the entire matrix is used.
    d - The damping factor, a float between 0 and 1. Defaults to .85.

    OUTPUTS:
    Use the eigenvalue solver in {scipy.linalg} to calculate the steady state of the PageRank algorithm.
    '''
    if N:
        A = data[:N,:N]
        n = N
    else:
        A = data
        n = datasize
        
    Am = A*np.eye(n)
    for i in xrange(n):
        if np.sum(Am,axis=1)[i] == 0:
            Am[i,:] += 1
            
    D = np.sum(Am,axis=1)
    K = Am.T/D
        
    B = d*K + ((1-d)/n)*np.ones((n,n))
    # Eigenvalue of largest magnitude, corresponding eigenvector
    evals, evecs = la.eig(B)
    return evecs[:,0]/np.sum(evecs[:,0])
```
