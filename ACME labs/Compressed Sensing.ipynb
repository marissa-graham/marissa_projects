```python
import numpy as np
from scipy import linalg as la
from cvxopt import matrix
from cvxopt import solvers
from matplotlib import pyplot as plt
%matplotlib inline

def l1Min(A,b):
    """Find the l1 minimum of x subject to Ax=b."""
    
    m = A.shape[0]
    n = A.shape[1]
    
    c = matrix(np.hstack((np.ones(n),np.zeros(n))))
    
    g1 = np.hstack((-np.eye(n),np.eye(n)))
    g2 = np.hstack((-np.eye(n),-np.eye(n)))
    G = matrix(np.vstack((g1,g2)))
    h = matrix(np.hstack(np.zeros(2*n)))
    
    Aprime = matrix(np.hstack((np.zeros((m,n)),A)))        
    b = matrix(b)
    
    # c is the objective function, G is the inequality constraint matrix,
    # h is its corresponding schtuff, then Ax=b constraint
    solvers.options['show_progress'] = False
    sol = solvers.lp(c,G,h,Aprime,b)
    return np.array(sol['x'])
    
def reconstruct(m):
    acme = 1-plt.imread('ACME.png')[:,:,0]
    np.random.seed(1337)
    A = np.random.randint(0,high=2,size=(m,32**2))
    b = A.dot(acme.flatten())
    rec_acme = np.array(l1Min(A,b)[1024:]).reshape(32,32)
    plt.imshow(rec_acme)
    plt.show()
    
    return la.norm(rec_acme-acme)

def problem2():
    acme = 1-plt.imread('ACME.png')[:,:,0]
    plt.imshow(acme)
    plt.show()
    
    print 200
    a = reconstruct(200)
    print 250
    b = reconstruct(250)
    print 260
    reconstruct(260)
    print 261
    reconstruct(261)
    print 262
    reconstruct(262)
    print 263
    reconstruct(263)
    print 270
    c = reconstruct(270)
    print '2-norm distance for 200 measurements: ' + str(a)
    print '2-norm distance for 250 measurements: ' + str(b)
    print '2-norm distance for 270 measurements: ' + str(c)
    
problem2()
```
