```python
# Linked lists
class Node():
    def __init__(self, data):
        self.next = None
        self.previous = None
        self.value = data

class SLinkedList(object):
    def __init__(self):
        self.head = None
        self.tail = None
        self.counter = 0
    
    def size(self):
        return self.counter

    def find(self, index):
        nfind = self.head
        nprev = None
        count = 0
        if index >= self.counter:
            raise IndexError
        else:
            while count < index and nfind.next:
                count += 1
                nprev = nfind
                nfind = nfind.next
            return nprev, nfind
    
    def insert(self, index, data):
        n = Node(data)
        if index > self.counter:
            raise IndexError
        if index == 0:
            n.next = self.head
            self.head = n
            if self.counter == 0:
                self.tail = self.head
        elif index == self.counter:
            self.tail.next=n
            self.tail=n
        else:
            nprev, nindex = self.find(index)
            n.next = nindex
            nprev.next = n
        self.counter += 1
        
    def clear(self):
        self.head = None
        
    def __str__(self):
        return '[' + ','.join(map(str, iter(self))) + ']'
    
    def __iter__(self):
        temp = self.head
        while temp:
            yield temp.value
            temp = temp.right
            
    # Problem 1
    def remove(self, index):
        if index >= self.counter:
            raise IndexError
        if self.counter == 1:
            clear(self)
        if index == 0:
            self.head = self.head.next
        elif index == self.counter-1:
            nprev, n = find(self, index)
            self.tail = nprev
            nprev.next = None
        else:
            nprev, n = self.find(index)
            nprev.next = n.next
        self.counter -= 1

# Problem 2
class MyStack(SLinkedList):
    def add(self, data):
        SLinkedList.insert(self, self.counter, data)
    def remove(self, data):
        SLinkedList.remove(self, self.counter-1)

class MyQueue(SLinkedList):
    def add(self, data):
        SLinkedList.insert(self, self.counter, data)
    def remove(self, data):
        SLinkedList.remove(self, 0)
```

```python
# Hash Tables
def HashFunction(string, tablesize, prime=5417):
        index = (len(string)*prime) % tablesize
        return index

class HashTable(object):
    def __init__(self, hsize, hashfunc):
        self.hashtable = [None]*hsize
        self.elsize = 0
        self.hsize = hsize
        self._hash = hashfunc
        
    def linear_find(self, data):
        index = self._hash(data, self.hsize)
        for i in xrange(self.hsize):
            newindex = (index + i) % self.hsize
            print 'Searching' + str(newindex)
            if self.hashtable[newindex] == data:
                return True
        return False
        
    def linear_insert(self, data):
        index = self._hash(data, self.hsize)
        if self.elsize <= self.hsize:
            newindex = index
            while self.hashtable[newindex] != None:
                newindex = (newindex + 1) % self.hsize
            self.hashtable[newindex] = data
            self.elsize +=1
        else:
            print 'Sorry, this hash table is full and the data cannot be added.'
            
    def chain_insert(self, data):
        index = self._hash(data, self.hsize)
        if self.hashtable[index] is None:
            self.hashtable[index] = [data]
        else:
            self.hashtable[index].append(data)
        self.elsize += 1
        self.resize()
        
    def chain_find(self, data):
        index = self._hash(data, self.hsize)
        for i in self.hashtable[index]:
            if i == data:
                return True
        return False
    
    def load(self):
        loadfactor = float(self.elsize)/self.hsize
        return loadfactor
    
    def resize(self):
        loadfactor = self.load()
        old_size = self.hsize
        if loadfactor > .75:
            while loadfactor > .33:
                self.hsize += 1
                loadfactor = self.load()
            newhash = HashTable(self.hsize, HashFunction)
            for i in self.hashtable:
                if i is not None:
                    for j in i:
                        index = self._hash(j, self.hsize)
                        if newhash.hashtable[index] is None:
                            newhash.hashtable[index] = [j]
                        else:
                            newhash.hashtable[index].append(j)
            self.hashtable = newhash.hashtable
                    
                         
linear_hash = HashTable(11, HashFunction)
felines = ['lion', 'tiger', 'cheetah', 'cougar', 'colocolo', 'cat']
for i in range(len(felines)):
    linear_hash.linear_insert(felines[i])
print linear_hash.hashtable
linear_hash.linear_insert('clouded leopard')
linear_hash.linear_insert('jaguar')
print linear_hash.hashtable
# 'clouded leopard' and 'jaguar' replaced 'lion' and 'cougar', respectively
# before I enabled linear probing. Also, elsize says there are 8 elements
# in the table which is incorrect; there are only six.
# After I enabled linear probing, clouded leopard was sent to index 10 and 
# jaguar was sent to index 0

chain_hash = HashTable(11, HashFunction)
for i in range(len(felines)):
    chain_hash.chain_insert(felines[i])
chain_hash.chain_insert('clouded leopard')
chain_hash.chain_insert('jaguar')
print chain_hash.hashtable
print chain_hash.hsize

chain_hash.chain_insert('siberian tiger')
chain_hash.chain_insert('bengal tiger')
chain_hash.chain_insert('liger')
print chain_hash.hsize
print chain_hash.hashtable
```
