```python
import scipy.stats as stats
import scipy as sp
import math as m
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np

def problem1():
    invg_rv1 = stats.invgamma(1, scale=1)
    invg_rv2 = stats.invgamma(1, scale=3)
    invg_rv3 = stats.invgamma(0.5, scale=5)
    x = np.linspace(-5,25,250)
    y1 = invg_rv1.pdf(x)
    y2 = invg_rv2.pdf(x)
    y3 = invg_rv3.pdf(x)
    plt.subplot(311)
    plt.plot(x,y1)
    plt.subplot(312)
    plt.plot(x,y2)
    plt.subplot(313)
    plt.plot(x,y3)

def problem2():
    x = np.linspace(0,1,300)
    prior = stats.beta(1,1)
    plt.subplot(211)
    plt.title('5 coin flips')
    y = prior.pdf(x)
    y12 = stats.beta(1,2).pdf(x)
    y22 = stats.beta(2,2).pdf(x)
    y32 = stats.beta(3,2).pdf(x)
    y33 = stats.beta(3,3).pdf(x)
    y43 = stats.beta(4,3).pdf(x)
    
    y21 = stats.beta(2,1).pdf(x)
    y31 = stats.beta(3,1).pdf(x)
    y41 = stats.beta(4,1).pdf(x)
    y42 = stats.beta(4,2).pdf(x)
    y52 = stats.beta(5,2).pdf(x)
    y62 = stats.beta(6,2).pdf(x)
    y63 = stats.beta(6,3).pdf(x)
    y64 = stats.beta(7,3).pdf(x)
    y74 = stats.beta(8,3).pdf(x)
    y84 = stats.beta(9,3).pdf(x)
    plt.plot(x,y,x,y12,x,y22,x,y32,x,y33,x,y43)
    plt.subplot(212)
    plt.plot(x,y,x,y21,x,y31,x,y41,x,y42,x,y52,x,y62,x,y63,x,y64,x,y74,x,y84)
    plt.title('10 coin flips')
    plt.show()
    print 'In the first distribution, the peak starts to grow larger as we flip the coin repeatedly but it stays relatively centered. The second trial is obviously biased towards heads so the distribution ends up skewing heads.'

def problem3():
    data = []
    with open('normalmean.dat', 'r') as f:
        for line in f:
            data.append(line.strip())
    data = np.array(data).astype(float)
    
    variance = np.var(data)
    mean = np.mean(data)

    x = np.linspace(40,120,300)
    y1 = stats.norm(loc=82,scale=6).pdf(x)
    y2 = stats.norm(loc=68,scale=7).pdf(x)
    y3 = stats.norm(loc=88,scale=4).pdf(x)

    def postnorm(a, b, dat=data, var=variance):
        mu_post = (a/b + sum(data)/var) / (1/b + float(len(dat))/var)
        sigsquared_post = (1/b + float(len(dat))/var)**(-1)
        sig_post = m.sqrt(sigsquared_post)
        normpost = stats.norm(loc=mu_post, scale=sig_post)
        return normpost

    prior = stats.invgamma(a=2, scale=100).pdf(x)
    alpha_post = 2 + len(data)/2.0
    beta_post = 100 + sum((data - mean)**2)/2.0
    post = stats.invgamma(a=alpha_post, scale=beta_post).pdf(x)

    plt.subplot(221)
    plt.plot(x, y1, x, postnorm(82,36).pdf(x))
    plt.title('Reasonably optimistic prior')
    print 'Reasonably optimistic prior- known variance:'
    print '95% confidence interval for mu is ' + str(postnorm(82,36).interval(.95))

    plt.subplot(222)
    plt.plot(x, y2, x, postnorm(68,49).pdf(x))
    plt.title('Pessimistic prior')
    print 'Pessimistic prior- known variance:'
    print '95% confidence interval for mu is ' + str(postnorm(68,49).interval(.95))

    plt.subplot(223)
    plt.plot(x, y3, x, postnorm(88,16).pdf(x))
    plt.title('Optimistic prior')
    print 'Optimistic prior- known variance:'
    print '95% confidence interval for mu is ' + str(postnorm(88,16).interval(.95))

    plt.subplot(224)
    plt.plot(x, prior, x, post)
    plt.title('Known mean')

    plt.tight_layout()
    plt.show()
```
