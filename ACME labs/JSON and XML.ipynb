```python
import json
import string
import datetime
import xml.etree.ElementTree as et

class DateEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, datetime.datetime):
            return {u'year':obj.year,u'month':obj.month,u'day':obj.day}
        return json.JSONEncoder.default(self, obj)

def date_decoder(dct):
    # try to decode any value that looks like a date
    # datetime has year, month, day, hour, minute, second, microsecond, tzinfo
    for i, k in dct.iteritems():
        try:
            dct[i] = k
        except:
            continue
    return dct

def prob1():
    today = datetime.datetime(1,4,8)
    a = json.dumps(today,cls=DateEncoder)
    print a
    print json.loads(a, object_hook=date_decoder)

def prob2():
    # Tags: author, description, publish_date, price
    # Find the most expensive book in the list
    # Find the number of books published before May 1, 2000
    # Which books reference Microsoft in their descriptions
    
    maxprice = 0
    priciest = ""
    numpublished = 0
    microsoftref = []
    f = et.iterparse('books.xml')
    for event, tag in f:
        if tag.tag == 'title':
            title = tag.text
        if tag.tag == 'price' and float(tag.text)>maxprice:
            maxprice = float(tag.text)
            priciest = title
        if tag.tag == 'publish_date':
            before = False
            date = string.split(tag.text,'-')
            if int(date[0])<=2000:
                before = True
            if int(date[0])==2001 and int(date[1])<5:
                before = True
            if before:
                numpublished += 1
        if tag.tag == 'description':
            if string.find(tag.text,'Microsoft') > 0:
                microsoftref.append(title)
    print "Most expensive book: \n\t" + priciest
    print "Number of books published before May 1, 2000: " + str(numpublished)
    print "Books which reference Microsoft in their descriptions:"
    for i in xrange(len(microsoftref)):
        print "\t" + microsoftref[i]
```
