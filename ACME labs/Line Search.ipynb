```python
import numpy as np
import scipy.linalg as la
from scipy.optimize import line_search
from scipy.optimize import leastsq
from matplotlib import pyplot as plt
%matplotlib inline

def newton1d(f, df, ddf, x, niter=10):
    '''
    Perform Newton's method to minimize a function from R to R
    
    Parameters
    ----------
    f: callable function object, the objective function
    df: callable function object, first derivative
    ddf: callable function object, the second derivative
    x: float, the initial guess
    niter: number of iterations
    
    Returns
    -------
    min: the approximated minimizer
    '''
    xold = x
    for i in xrange(niter):
        xnew = xold - df(xold)/ddf(xold)
        xold = xnew
    return xnew

def backtracking(f, slope, x, p, a=1, rho=.9, c=10e-4):
    '''
    Perform a backtracking line search to satisfy the Armijo condition.
    
    Parameters
    ----------
    f: The objective function
    slope: the value of grad(f)^T/p
    x: current iterate; ndarray of shape (n,)
    p: current search direction; ndarray of shape (n,)
    a: initial step length (set to 1 in Newton and quasi-Newton methods)
    rho: a number in (0,1)
    c: a number in (0,1)
    
    Returns
    -------
    alpha: computed step size satisfying the Armijo condition
    '''
    
    while f(x+alpha*p) <= f(x) + c*a*slope:
        a = rho*a
    return a

def gradientDescent(f, df, x, niter=10):
    '''
    Minimize a function using gradient descent.
    Parameters
    ----------
    f : callable function object
        A differentiable real-valued function
    df : callable function object
        The gradient of the function
    x : ndarray of shape (n,)
        The initial point
    niter : integer
        The number of iterations to run.
    Returns
    -------
    pts: list of ndarrays
    The sequence of points generated
    '''
    xk = [x]
    for i in xrange(niter):
        pk = -df(xk[-1])
        slope = np.dot(df(xk[-1]).T,pk)
        alpha = backtracking(f, slope, xk[-1], pk)
        xk.append(xk[-1] + alpha*pk)
    return xk

def newtonsMethod(f, df, ddf, x, niter=10):
    '''
    Minimize a function using Newton's method.
    
    Parameters
    ----------
    f : callable function object
        Real-valued, twice-differentiable function
    df : callable function object
        The gradient of the function
    ddf : callable function object
        The Hessian of the function
    x : ndarray of shape (n,)
        The initial point
    niter : integer
        The number of iterations
        
    Returns
    -------
    pts : list of ndarrays
    The sequence of points generated
    '''
    pk = la.solve(ddf(xk),-df(xk))
    xk = x
    for i in xrange(niter):
        
        slope = np.dot(df(xk).T,pk)
        alpha = backtracking(f, slope, xk, pk)
        xk = xk + alpha*pk
        
def gaussNewton(f, df, jac, r, x, niter=10):
    '''
    Solve a nonlinear least squares problem with Gauss-Newton method.
    Parameters
    ----------
    f : callable function object
    The objective function
    df : callable function object
    The gradient of f
    jac : callable function object
    The jacobian of residual vector
    r : callable function object
    The residual vector
    x : ndarray of shape (n,)
    The initial point12 Lab 1. Line Search Algorithms
    niter : integer
    The number of iterations
    Returns
    -------
    min : ndarray of shape (n,)
    The minimizer
    '''
    k = 0
    thing = np.dot(jac(x).T,r(x))
    ak = [0]
    while thing.all() != 0 and k < niter:
        k += 1
        jacx = jac(x)
        thing = np.dot(jac(x).T, r(x))
        pk = la.solve(np.dot(jac(x).T,jac(x)),-thing)
        ak = line_search(f, df, x, pk)[0]
        if ak == None:
            ak = 0
        x = x + ak*pk
    return x

def check_gaussNewton():
    t = np.arange(10)
    y = 3*np.sin(0.5*t) + 0.5*np.random.randn(10)
    def model(x,t):
        return x[0]*np.sin(x[1]*t)
    def residual(x):
        return model(x,t)-y
    def jac(x):
        ans = np.empty((10,2))
        ans[:,0] = np.sin(x[1]*t)
        ans[:,1] = x[0]*t*np.cos(x[1]*t)
        return ans
    def objective(x):
        return .5*(residual(x)**2).sum()
    def grad(x):
        return jac(x).T.dot(residual(x))
    x0 = np.array([2.5,0.6])
    x = gaussNewton(objective, grad, jac, residual, x0, niter=10)
    domain = np.linspace(0,10,100)
    plt.plot(t, y, '*')
    plt.plot(domain, 3*np.sin(.5*domain), '--')
    plt.plot(domain, x[0]*np.sin(x[1]*domain))
    plt.show()
    
def problem5():
    # Start with the first 8 decades of data
    years1 = np.arange(8)
    pop1 = np.array([3.929, 5.308, 7.240, 9.638, 12.866, 17.069, 23.192, 31.443])
    
    # Now consider the first 16 decades
    years2 = np.arange(16)
    pop2 = np.array([3.929, 5.308, 7.240, 9.638, 12.866, 17.069, 23.192, 31.443, 38.558, 50.156, 62.948, 75.996, 91.972, 105.711, 122.775, 131.669])
    
    # Set up the first model
    def model1(x, years1):
        return x[0]*np.exp(x[1]*(years1+x[2]))
    def residual1(x):
        return model1(x, years1) - pop1
    c1 = [150.0, 0.4, 2.5]    
    x1 = leastsq(residual1, c1)[0]
    
    # Set up the second model
    def model2(x, years2):
        return x[0]/(1+np.exp(-x[1]*(years2+x[2])))
    def residual2(x):
        return model2(x, years2) - pop2
    c2 = [100, 1.0, 2.0]
    x2 = leastsq(residual2, c2)[0]
    
    plt.plot(years1, pop1, '*')
    plt.plot(years1, model1(x1, years1))
    plt.show()
    
    plt.plot(years2, pop2, '*')
    plt.plot(years2, model2(x2, years2))
    plt.show()
```
