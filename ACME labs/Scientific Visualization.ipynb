```python
from matplotlib import pyplot as plt
import numpy as np

def problem1():
    x1 = np.array([10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0])
    y1 = np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68])
    
    x2 = x1
    y2 = np.array([9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74])
    
    x3 = x1
    y3 = np.array([7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73])
    
    x4 = np.array([8.0, 8.0, 8.0, 8.0, 8.0, 8.0, 8.0, 19.0, 8.0, 8.0, 8.0])
    y4 = np.array([6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89])
    
    plt.subplot(221)
    plt.scatter(x1, y1)
    
    plt.subplot(222)
    plt.scatter(x2, y2)
    
    plt.subplot(223)
    plt.scatter(x3, y3)
    
    plt.subplot(224)
    plt.scatter(x4, y4)
    
    plt.suptitle("Anscombe's quartet")
    plt.show()
    # Plot the four data sets
    
    observations = 'Even though the statistical properties of these sets are the same, visually, each set looks very different. One looks like a typical scatter plot, one looks like a parabola, one looks like a straight line with one outlier, and the last is a vertical line with one outlier'
    print observations
    
def problem2():
    x = np.array([1000,2000,4000,8000])
    y = np.array([11.2,45.8,185,753])
    
    logx = np.log(x)
    logy = np.log(y)
    
    plt.subplot(311)
    plt.plot(logx, logy)
    plt.title('log-log scale')
    
    plt.subplot(312)
    plt.plot(x, logy)
    plt.title('log-lin scale')
    
    plt.subplot(313)
    plt.plot(logx, y)
    plt.title('lin-log scale')
    
    plt.show()
    
    easiest_to_understand = 'lin-log, since it better shows that the data is increasing at a polynomial rate.'
    print easiest_to_understand

def problem3():
    x = np.linspace(0,10,500)
    y = (np.sin(x))/(x+1)
    plt.fill_between(x, 0, y, where=y>0, color='blue')
    plt.fill_between(x, 0, y, where=y<0, color='red')
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.title("My Original Plot")

    indices = np.zeros_like(x, dtype=bool)
    for i in xrange(1, y.size-1):
        if (y[i-1]-y[i])*(y[i]-y[i+1])<0:
            indices[i]=True
        
    plt.scatter(x[indices], .5*y[indices], marker='^')
    plt.xlim([0,10])
    plt.plot(x, y, ':')
    plt.grid()
    plt.show()
    
    plt.plot(x,y)
    plt.scatter(x[indices], y[indices])
    plt.xlim([0,10])
    plt.ylim([-0.25,0.5])
    axis = plt.gca()
    axis.spines['right'].set_visible(False)
    axis.spines['top'].set_visible(False)
    plt.xticks(np.arange(0,11,5))
    plt.yticks(np.array([-0.25,0,0.25,0.5]))
    plt.title('Simplified Plot')
    plt.show()
```
