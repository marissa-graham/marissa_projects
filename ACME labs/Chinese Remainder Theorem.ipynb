```python
def egcd(a, b):
    """Return the gcd of a and b, as well as integers x
    and y such that a*x+b*y=gcd(a,b).
    """
    x,y, u,v = 0,1, 1,0
    while a != 0:
        q, r = b//a, b%a
        m, n = x-u*q, y-v*q
        b,a, x,y, u,v = a,r, u,v, m,n
    gcd = b
    return gcd, x, y

def prod(iterable):
    """Return the product of the elements in an iterable."""
    prod = 1
    for i in xrange(len(iterable)):
        prod *= iterable[i]
    return prod

def lagrange(x,m):
    """Implement the Lagrange Decomposition Algorithm
    for integers in Python.
    
    INPUT:
    x - integer
    m - tuple of pairwise relatively prime integers
    
    OUTPUT:
    xtuple - a tuple (x1,...,xn) such that x is congruent to 
        x1+...+xn mod prod(m) and xi congruent to 0 for all mj
        whenever i != j.
    """
    n = len(m)
    M = prod(m)
    L = ()
    fbar = ()
    xtuple = ()
    for i in xrange(n):
        pi = M/m[i]
        st = egcd(pi,m[i])
        L += (st[1]*pi,)
        fbar += (x*st[1]*pi,)
        xtuple += (x*st[1]*pi % M,)
    return xtuple

def CRT_newton(a,m):
    """Use the Chinese remainder theorem to find an integer x
    such that x is congruent to a[i] mod m[i] for all i, using
    the Newton-Garner formula.
    """
    n = len(m)
    b = a[0]
    s = egcd(m[0],m[1])[1]
    x = a[0]
    for i in xrange(1,n):
        mprod = prod(m[0:i])
        s = egcd(mprod,m[i])[1]
        b = s*(a[i]-x)
        x += b*mprod
    return x % prod(m)
    

def CRT_lagrange(a,m):
    """Use the Chinese remainder theorem to find an integer x
    such that x is congruent to a[i] mod m[i] for all i,
    using the Lagrange decomposition.
    """
    n = len(m)
    M = prod(m)
    x = 0
    for i in xrange(n):
        pi = M/m[i]
        st = egcd(pi,m[i])
        x += st[1]*pi*a[i]
    return x % M
```
