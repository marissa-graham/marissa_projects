```python
import numpy as np
from matplotlib import pyplot as plt

def plot_complex(f, xbounds, ybounds, res=401, colormap='hsv'):

    '''Plot the complex function f.
    INPUTS:
    f - A function handle. Should represent a function from C to C.
    xbounds - A tuple (xmin, xmax) describing the bounds on the real part of the domain.
    ybounds - A tuple (ymin, ymax) describing the bounds on the imaginary part of the domain.
    res - A scalar that determines the resolution of the plot. Defaults to 401.
    
    OUTPUTS:
    graph of the function f(z)
    '''
    x = np.linspace(xbounds[0],xbounds[1],res)
    y = np.linspace(ybounds[0],ybounds[1],res)
    X, Y = np.meshgrid(x,y)
    Z = X + 1j*Y
    plt.pcolormesh(X, Y, np.angle(f(Z)), cmap=colormap, vmin=-np.pi, vmax=np.pi)
    plt.show()    
    
f = lambda z:z**4 + 8*z**3 + 3*z**2 + 8*z + 3
a = 8
plot_complex(f, (-a,a), (-a,a))
    
def problem1():
    print "\nProblem 1:\n"
    f = lambda z:z
    plot_complex(f, (-1,1), (-1,1))
    g = lambda z:np.sqrt(z**2+1)
    plot_complex(g, (-3,3), (-3,3))
    
def problem2():
    print "\nProblem 2:\n"
    for i in xrange(2,5):
        print "Plot of z^" + str(i) + ':'
        f = lambda z:z**i
        plot_complex(f, (-1,1), (-1,1))
    print "Plot of z^3-iz^4-3z^6:"
    g = lambda z:z**3-1j*z**4-3*z**6
    plot_complex(g, (-1,1), (-1,1))
    print 'The zeros of the function seem to be the points where all the colors touch each other.'
    print 'The multiplicity of the zero is found by counting the rainbows around the singularity'
    print 'This function has a zero at the origin, as does z^3, but also has three additional zeros, so it looks like z^3 near the origin.'

def problem3():
    print "\nProblem 3:\n"
    for i in xrange(1,4):
        print 'Plot of z^-' + str(i) + ':'
        f = lambda z:1/z**i
        plot_complex(f, (-1,1), (-1,1))
    g = lambda z:z**2 + 1j/z + 1/z**3
    print 'Plot of z^2 + i/z + z^-3'
    plot_complex(g, (-1,1), (-1,1))
    print 'The last function and z^-3 behave very similarly at the origin.'
    print 'It seems that poles behave similarly to zeros, but the colors circle around in the opposite direction as a root.'

def problem4():
    print '\nProblem 4:\n'
    f = lambda z:np.exp(z)
    print 'Plot of e^z:'
    plot_complex(f, (-8,8), (-8,8))
    print 'e^z does not have any roots or poles.'
    f = lambda z:np.tan(z)
    print 'Plot of tan(z):'
    plot_complex(f, (-8,8), (-8,8))
    print 'tan(z) has zeros at points 2kpi, and poles at 2kpi+pi/2, each with order 1.'
    f = lambda z:(16*z**4+32*z**3+32*z**2+16*z+4)/(16*z**4-16*z**3+5*z**2)
    print '\nPlot of funky rational function:'
    plot_complex(f, (-1,1), (-1,1))
    print 'We have two zeros of order two, one pole of order two, and two poles of order one.'

def problem5():
    print '\nProblem 5:\n'
    f = lambda z:-2*z**7 + 2*z**6 - 4*z**5 + 2*z**4 - 2*z**3 - 4*z**2 + 4*z - 4
    g = lambda z:z**7 + 6*z**6 - 131*z**5 - 419*z**4 + 4906*z**3 - 131*z**2 - 420*z + 4900
    print 'Plot of -2z^7+2z^6-4z^5+2z^4-2z^3-4z^2+4z-4:'
    plot_complex(f, (-2,2), (-2,2))
    print 'We have seven zeros of order one, which is all of them by the Fundamental Theorem of Algebra.'
    print '\nPlot of z^7+6z^6-131z^5-419z^4+4906z^3-131z^2-420z+4900'
    n=15
    plot_complex(g, (-n,n), (-n,n))
    print 'We have two zeros of order two and three zeros of order one, which adds to seven, so by the Fundamental Theorem of Algebra we have all of them.\n'

def problem6():
    print '\nProblem 6:\n'
    
    f = lambda z:np.sin(1.0/(100.0*z))
    n = 1.0
    print 'Plot of f on [-1,1]x[-1,1]:'
    plot_complex(f, (-n,n), (-n,n))
    n = .005
    print 'The function looks a lot like 1/z, with a simple pole at zero.'
    print '\nPlot of f on [-.01,.01]x[-.01,.01]:'
    plot_complex(f, (-n,n), (-n,n), res=500)
    print 'But actually we have an essential singularity, with infinite order as we approach zero (as in e^1/z).'

    g = lambda z:z+1000*z**2
    print '\nPlot of z+1000z^2 on [-1,1]x[-1,1]:'
    plot_complex(g, (-1,1),(-1,1))
    n = .002
    print 'We appear to have a single zero of order two. But solving algebraically gives a zero of order one at zero and a zero of order one at -1/1000.\n'
    print 'Plot of z+1000z^2 on a more appropriate domain:'
    plot_complex(g, (-n,n), (-n,n))
    print 'We can now see the two distinct zeros.'
    
def problem7():
    print '\nProblem 7\n'
    f = lambda z:np.sqrt(z)
    g = lambda z:-np.sqrt(z)
    print 'Plot of sqrt(z):'
    plot_complex(f, (-1,1), (-1,1))
    print 'Plot of -sqrt(z):'
    plot_complex(g, (-1,1), (-1,1))
    print 'The two plots look very similar, but in each, we lost half the colors (it has been stretched out).'
    
def checklab():
    problem1()
    problem2()
    problem3()
    problem4()
    problem5()
    problem6()
    problem7()
```
