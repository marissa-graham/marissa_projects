```python
import numpy as np
from scipy import linalg as la
from matplotlib import pyplot as plt
import sympy as sy
from scipy.misc import factorial
from IPython.display import display, Math, Latex
%matplotlib inline

# DON'T TURN IN WITHOUT TAKING OUT COMMENTS AND %MATPLOTLIB INLINE

##############  Problem 1  ##############
def prob1():
    '''
    Randomly perturb w_coeff by replacing each coefficient a_i with a_i*r_i, where
    r_i is drawn from a normal distribution centered at 1 with varience 1e-10.
    
    Plot the roots of 100 such experiments in a single graphic, along with the 
    roots of the unperturbed polynomial w(x)
    
    Using the final experiment only, estimate the relative and absolute 
    condition number (in any norm you prefer).
    
    Return:
    Should display graph of all 100 perturbations.
    Should print values of relative and absolute condition.
    '''
    w_coeffs = np.array([1, -210, 20615, -1256850, 53327946, -1672280820, 40171771630, -756111184500, 11310276995381, -135585182899530, 1307535010540395, -10142299865511450, 63030812099294896, -311333643161390640, 1206647803780373360, -3599979517947607200, 8037811822645051776, -12870931245150988800, 13803759753640704000, -8752948036761600000, 2432902008176640000])
    w_roots = np.arange(1,21)
    
    for i in xrange(100):
        perturb = np.random.normal(loc=1.0,scale=1e-10,size=21)
        perturbed_coeffs = w_coeffs*perturb
        perturbed_roots = np.zeros(21)
        perturbed_roots = np.roots(np.poly1d(perturbed_coeffs))
        
        plt.scatter(perturbed_roots.real,perturbed_roots.imag, s=1)
    
    plt.scatter(w_roots,np.zeros(20),s=20)
    plt.show()
    
    w_roots = np.sort(w_roots)
    perturbed_roots = np.sort(perturbed_roots)
    absolute = la.norm(perturbed_roots-w_roots,np.inf)/la.norm(perturbed_coeffs-w_coeffs,np.inf)
    relative = la.norm(w_coeffs,np.inf)/la.norm(w_roots,np.inf)*absolute
    print 'Absolute condition number: ', absolute
    print 'Relative condition number: ', relative
    
print 'Problem 1:\n\n'
prob1()

##############  Problem 2  ##############
print '\n\nProblem 2:\n\n'
def eig_condit(M):
    '''
    Approximate the condition number of the eigenvalue problem at M.

    INPUT:
    M - A 2-D NumPy array, representing a matrix.

    RETURN:
    A tuple containing approximations to the absolute and 
    relative condition numbers of the eigenvalue problem at M.
    '''
    eigs = la.eig(M)[0]
    perturb = np.random.normal(0,1e-10,M.shape)+np.random.normal(0,1e-10,M.shape)*1j
    eigsp = la.eig(M+perturb)[0]
    absolute = la.norm(eigs-eigsp)/la.norm(perturb)
    relative = absolute*la.norm(M)/la.norm(eigs)
    return (absolute,relative)   
    
print "Test Case: M=2x2 identity: ",str(eig_condit(np.array([[1,0],[0,1]])))

print "Example with large condition number: "
M = np.array([[1.0,0.0],[1e10,1.0]])
print M
print "Condition value: ",str(eig_condit(M)), '\n'


#   1 pt extra credit
def plot_eig_condit(x0=-100, x1=100, y0=-100, y1=100, res=10):
    '''
    Plot the condition number of the eigenvalue problem on [x0, x1]x[y0,y1].

    Specifically, use plt.pcolormesh to plot the relative condition 
    number of the eigenvalue problem at [[1,x],[y,1]] on this domain.

    The variable `res' should be the number of sample points taken 
    along each axis, for a total of `res'**2 points in the plot.
    '''
    x = np.linspace(x0,x1,num=res)
    y = np.linspace(y0,y1,num=res)
    X, Y = np.meshgrid(x,y)
    Z = np.zeros((res,res))
    for i in xrange(res):
        for j in xrange(res):
            Z[i,j] = eig_condit(np.array([[1,x[i]],[y[j],1]]))[1]
    plt.pcolormesh(X,Y,Z)
    plt.show()
       
for i in [10,50,100,200,400]:
    plot_eig_condit(res=i)

##############  Problem 3  ##############
def integral(n):
    '''
    RETURN I(n)
    '''
    e = np.exp(1)
    if n < 0:
        print 'No fair using negative numbers. Try again.'
    if n == 0:
        return (e-1)/e
    if n == 1:
        return 1/e
    if n > 1:
        return (-1)**n*sy.subfactorial(n)+(-1)**(n+1)*factorial(n)/e
       
print '\n\nProblem 3:\n\n'
print 'Proof that 0 <= I(n) <= 1:\n'
print 'We know that I(n)>=0 because the integrand is nonnegative. Then the integral is bounded by 1 times the integrand evaluated at 1, which is 1. So 0 <= I(n) <= 1.'
print "Just so you know, I did try to get this to display in LaTeX. I couldn't get it to work, but I did try."
        
actual_int = np.array([0.367879441171, 0.145532940573, 0.0838770701034, 0.0590175408793, 0.0455448840758, 0.0370862144237, 0.0312796739322, 0.0270462894091, 0.023822728669, 0.0212860390856, 0.0192377544343])
print '\nn    integral(n)    Actual value'
k = 0
for i in [1,5,10,15,20,25,30,35,40,45,50]:
    print i,' ', integral(i), '  ', actual_int[k]
    k += 1
print '\nFor small values of n, our integral() function is very close to the actual value. Once we hit n=20, n! becomes extremely large, and we begin experiencing overflow errors and instability, and everything blows up.'
```
