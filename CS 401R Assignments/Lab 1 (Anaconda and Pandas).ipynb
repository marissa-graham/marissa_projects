```python
from PIL import Image, ImageDraw
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

im = Image.new('RGB', (512,288), color='white')
draw = ImageDraw.Draw(im)

for i in xrange(250):
    x, y = np.random.randint(512,size=2), np.random.randint(288,size=2)
    ang, c = 360.0*np.random.rand(2), tuple(np.random.randint(0,256,size=3))
    draw.arc((x[0],y[0])+(x[1],y[1]), ang[0], ang[1], fill=c)
    
plt.imshow(np.asarray(im))
```
