```python
import csv
import math
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline

n = 56
responders = [[]]*n
with open('gendersurvey.csv') as data:
    i = 0
    datareader = csv.reader(data, delimiter=',')
    for line in datareader:
        responders[i].append(line)
        i += 1

responders = responders[0]

def prod(iterable):
    return reduce(operator.mul, iterable, 1)

gender = []
major = []
wherefrom = []
maritalstatus = []
normalmen = []
normalwomen = []
for i in xrange(1,len(responders)):
    gender.append(responders[i][2])
    major.append(responders[i][3])
    wherefrom.append(responders[i][4])
    maritalstatus.append(responders[i][6])
    men = 1
    women = 1
    for j in xrange(7):
        women *= float(responders[i][10+j])
        men *= float(responders[i][23+j])
    normalwomen.append(math.log(women))
    normalmen.append(math.log(men))

men1 = []
women1 = []
stem1 = []
arts1 = []
socsci1 = []
single1 = []
dating1 = []
married1 = []
men2 = []
women2 = []
stem2 = []
arts2 = []
socsci2 = []
single2 = []
dating2 = []
married2 = []
for i in xrange(n-1):
    if gender[i] == 'Male':
        men1.append(normalmen[i])
        men2.append(normalwomen[i])
    if gender[i] == 'Female':
        women1.append(normalmen[i])
        women2.append(normalwomen[i])
    if major[i][:4] == 'STEM':
        stem1.append(normalmen[i])
        stem2.append(normalwomen[i])
    if major[i][:7] == 'Liberal':
        arts1.append(normalmen[i])
        arts2.append(normalwomen[i])
    if major[i][:6] == 'Social':
        socsci1.append(normalmen[i])
        socsci2.append(normalwomen[i])
    if maritalstatus[i] == 'Married':
        married1.append(normalmen[i])
        married2.append(normalwomen[i])
    if maritalstatus[i] == 'Single':
        single1.append(normalmen[i])
        single2.append(normalwomen[i])
    if maritalstatus[i][:6] == 'Dating':
        dating1.append(normalmen[i])
        dating2.append(normalwomen[i])
        
men1 = np.mean(np.array(men1))
men2 = np.mean(np.array(men2))
women1 = np.mean(np.array(women1))
women2 = np.mean(np.array(women2))
stem1 = np.mean(np.array(stem1))
stem2 = np.mean(np.array(stem2))
arts1 = np.mean(np.array(arts1))
arts2 = np.mean(np.array(arts2))
socsci1 = np.mean(np.array(socsci1))
socsci2 = np.mean(np.array(socsci2))
married1 = np.mean(np.array(married1))
married2 = np.mean(np.array(married2))
single1 = np.mean(np.array(single1))
single2 = np.mean(np.array(single2))
dating1 = np.mean(np.array(dating1))
dating2 = np.mean(np.array(dating2))

ind = np.arange(2)
width = .35
fix, ax = plt.subplots()
rects1 = ax.bar(ind, (men1,women1), width, color='k')
rects2 = ax.bar(ind+width, (men2,women2), width, color='m')
ax.set_title('Perceptions of same-gender touch by gender')
ax.set_xticks(ind+width)
ax.set_xticklabels(("Men's perceptions", "Women's perceptions"))
ax.legend((rects1[0], rects2[0]), ('Male same-gender touch', 'Female same-gender touch'),loc='lower right')

def autolabel(rects):
    # attach some text labels
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2., 1.05*height, int(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
plt.show()

ind = np.arange(3)
width = .35
fix, ax = plt.subplots()
rects1 = ax.bar(ind, (stem1, arts1, socsci1), width, color='k')
rects2 = ax.bar(ind+width, (stem2, arts2, socsci2), width, color='m')
ax.set_title('Perceptions of same-gender touch by major')
ax.set_xticks(ind+width)
ax.set_xticklabels(("STEM majors", "Arts majors", "Social Science majors"))
ax.legend((rects1[0], rects2[0]), ('Male same-gender touch', 'Female same-gender touch'),loc='lower right')

def autolabel(rects):
    # attach some text labels
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2., 1.05*height, int(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
plt.show()

ind = np.arange(3)
width = .35
fix, ax = plt.subplots()
rects1 = ax.bar(ind, (married1, single1, dating1), width, color='k')
rects2 = ax.bar(ind+width, (married2, single2, dating2), width, color='m')
ax.set_title('Perceptions of same-gender touch by marital status')
ax.set_xticks(ind+width)
ax.set_xticklabels(("Married", "Single", "Dating/Engaged"))
ax.legend((rects1[0], rects2[0]), ('Male same-gender touch', 'Female same-gender touch'),loc='lower right')

def autolabel(rects):
    # attach some text labels
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2., 1.05*height, int(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
plt.show()

print '\n\n\n\n'
```
