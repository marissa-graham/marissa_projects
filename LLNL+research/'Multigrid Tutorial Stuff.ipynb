```python
def prime_check(n):
    
    primes = np.ones(n)
    
    primes[0:2] = 0
    for i in xrange(2,int(np.sqrt(n))):
        primes[2*i::i] = 0
    
    return np.where(primes==1)
    
print prime_check(150)
```

```python
# Aliasing demonstration

n = 12

for k in xrange(1,n):
    print "Fourier mode #", k, " and alias ", 2*n-k
    plt.subplot(121)
    x = np.linspace(0,1,n+1)
    plt.plot(x,np.sin(k*np.pi*x))
    plt.subplot(122)
    plt.plot(x,np.sin((2.0*n-k)*np.pi*x))
    plt.show()
```

```python
def plot_evals(n, w):
    dom = np.linspace(0,1,300)
    for l in xrange(n/2,n):
        plt.plot(dom, np.abs(1-w*(np.sin(dom*np.pi/2.0)**2+np.sin(l*np.pi/(2*n))**2)))
    plt.axvline(x=0.5)
    plt.axhline(y=0)
    plt.axhline(y=0.5)
    plt.show()
    
for i in [0.35,0.5,2/3.,0.8,1.0]:
    print i
    plot_evals(16,i)
```
