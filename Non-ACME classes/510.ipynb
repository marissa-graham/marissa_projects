```python
# Homework 11

def arnoldi(A,b):
    H = np.zeros_like(A)
    Q = np.zeros_like(A)
    Q[:,0] = b/la.norm(b)
    done = False
    n = 0
    while not done:
        v = np.dot(A,Q[:,n])
        for j in xrange(n):
            H[j,n] = np.dot(Q[:,j],v)
            v -= H[j,n]*Q[:,j]
        H[n+1,n] = la.norm(v)
        if np.abs(H[n+1,n]) < 1e-16 or n >= A.shape[0]-2:
            done = True
        else:
            Q[:,n+1] = v/H[n+1,n]
            n += 1
    return Q[:,:n+2], H[:n+2,:n+1]

# 34.3b
dvals = np.array([1.0/np.sqrt(k+1) for k in xrange(64)])
A = np.diag(dvals)+np.diag(dvals[:-1],1)
evals, evecs = la.eig(A)
Q, H = arnoldi(A,np.ones(64))
plt.scatter(np.real(evals),np.imag(evals))
plt.show()
for i in xrange(30):
    print "n = "+ str(i+1)
    h = H[:i+1,:i+1]
    w, v = la.eig(h)
    plt.scatter(np.real(evals),np.imag(evals))
    plt.scatter(np.real(w),np.imag(w),c='red')
    plt.show()
```

```python
# Homework 9

def maketest(m):
    subsup = -1*np.ones(m-1)
    return np.diag(np.ones(m))+np.diag(subsup,1)+np.diag(subsup,-1)

# Algorithm 27.1: Power Iteration
def powerEig(A,v0,tol):
    """Compute the eigenvector of A corresponding to the 
    largest eigenvalue, using the initial unit vector v0."""
    i = 0 # Keep track of number of iterations
    v = v0
    error = 1.0
    lambd = 0.0
    while error > tol and i < 10000000:
        prevlambd = lambd
        w = np.dot(A,v)
        v = w/la.norm(w)
        lambd = np.dot(v,np.dot(A,v))
        error = np.abs(lambd - prevlambd)
        i += 1
    return v, lambd, i

A = maketest(10)
v0 = np.ones(10)
v0[::2] *= -1
v, lambd, i = powerEig(A,v0/la.norm(v0),10e-10)
eigs, V = la.eig(A)
# Note: positive multiples of a ones vector converge to the second largest
# eigenvalue. The v0 given converges to the correct value. More accuracy 
# than scipy gives in less than 38 iterations.

# Algorithm 27.2: Inverse Iteration
def inverseEig(A,v0,mu,tol):
    v = v0
    i = 0
    error = 1.0
    lambd = 0.0
    while error > tol and i < 1000000:
        prevlambd = lambd
        w = la.solve(A-mu*np.eye(A.shape[0]),v)
        v = w/la.norm(w)
        lambd = np.dot(v,np.dot(A,v))
        error = np.abs(prevlambd - lambd)
        i += 1
    return v, lambd, i

A = maketest(10)
v0 = np.ones(10)
v0[::2] *= -1
v, lambd, i = inverseEig(A,v0/la.norm(v0),25,10e-7)
eigs, V = la.eig(A)
# Learned the hard way to use lambd as the stopping condition and not 
# the norms of the eigenvectors. Which eigenvalue it converges to is 
# dependent on mu. Not predictable exactly which eigenvalue we will get,
# but it will be one of the closer ones to mu, it seems, and mu's that
# aren't close to any eigenvalue will take a lot longer to converge.

# Algorithm 27.3: Rayleigh Quotient iteration
def rayleighEig(A,v0,tol):
    v = v0
    lambd = np.dot(v,np.dot(A,v))
    error = 1.0
    i = 0
    while error > tol and i < 100000000:
        w = la.solve(A-lambd*np.eye(A.shape[0]),v)
        v = w/la.norm(w)
        prevlambd = lambd
        lambd = np.dot(v,np.dot(A,v))
        error = np.abs(prevlambd-lambd)
        i += 1
    return v, lambd, i

A = maketest(10)
v0 = np.ones(10)
v, lambd, i = rayleighEig(A,v0/la.norm(v0),10e-15)
eigs, V = la.eig(A)
# WOW GETTING, LIKE, MACHINE ACCURACY IN LESS THAN TEN ITERATIONS, MORE 
# LIKE FOUR, EVEN WHEN I TRY TO GIVE IT A BAD INITIAL GUESS THIS IS COOL 

# 28.2
def test_zeros(n):
    a1 = np.random.rand(n-1)
    a2 = np.random.rand(n)
    A = np.diag(a1,-1)+np.diag(a2)+np.diag(a1,1)
    Q, R = la.qr(A)
    # Generally, R is nonzero along the diagonal and the first and second
    # superdiagonals. We know this both experimentally and
    # because R is formed by dotting vectors from Q with A; zeros in A
    # are largely preserved, but only in certain locations.
    # Q is zero below the first subdiagonal. Examining the QR decomposition
    # algorithm, we note that this is because Q is initialized as A, so is
    # initially tridiagonal, but then loses the zeros in the upper half as
    # the projections of previous columns (which have nonzero entries there)
    # are subtracted.
    
    # By investigating the structure of R and Q, and the placement of the 
    # zeros, we see through investigating the matrix multiplication (and 
    # running experiments for random A) that the tridiagonal structure is
    # preserved.
    
    print R, '\n'
    print Q, '\n'
    RQ = np.dot(R,Q)
    RQ[np.abs(RQ)<1e-15] = 0
    print RQ
    
test_zeros(7)
```

```python
# Homework 3

def mgs(A):
    """
    Compute a reduced QR factorization of mxn A, m >= n, 
    using modified Gram-Schmidt orthogonalization.
    
    Output: mxn complex valued Q with orthonormal columns
    Triangular complex valued R, nxn
    """
    m, n = A.shape
    
    R = np.zeros((n,n))
    Q = A.astype(float) # Overwrite things
    
    for i in xrange(n):
        R[i,i] = la.norm(Q[:,i])
        Q[:,i] /= R[i,i]
        
        for j in xrange(i+1,n):
            R[i,j] = np.dot(Q[:,i],A[:,j])
            Q[:,j] = Q[:,j] - R[i,j]*Q[:,i]
    
    return Q, R

A = np.array([[1,2],[0,1],[1,0]],dtype=float)
Q, R = mgs(A)
q, r = la.qr(A)
print Q, '\n\n', q, '\n\n', R, '\n\n', r
```
